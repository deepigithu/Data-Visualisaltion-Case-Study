```python
import numpy as np
import pandas as pd 
import datetime as dt 
import matplotlib.pyplot as  plt  
import seaborn as sns 
```

```python
sales_data=pd.read_csv(r'C:\Users\Deepika\Downloads\PYTHON\SalesData.csv')
```

```python
sales_data
```

```python
#1
region_sales=sales_data.groupby(by='Region')[['Sales2015','Sales2016']].sum()
```

```python
region_sales.plot(kind='bar',title="Sales by Region in 2015 and 2016")
plt.show()
```

```python
#2
region_sales2016=sales_data.groupby(by='Region')['Sales2016'].sum()
```

```python
plt.pie(region_sales2016,autopct="%1.0f%%",labels=['Central','East','West'],shadow=True,colors=['r','g','b'])
plt.title("Contributing Factors  to the Sales for each Region in 2016")
plt.show()
```

```python
#3
sales_by_reti=sales_data.groupby(by=['Region','Tier'])[['Sales2015','Sales2016']].sum()
```

```python
sales_by_reti.plot(kind='bar',figsize=[10,5],title='Total Sales of 2015 and 2016 with respect to Region and Tier')
plt.show()
```

```python
#4
EastRegion=sales_data.loc[sales_data.Region=='East']
EastRegion
```

```python
east_sales=EastRegion.groupby(by=['Region','State'])[['Sales2015','Sales2016']].sum()
```

```python
east_sales.plot(kind='bar',figsize=[10,5],title='Comparison of Total Sales in 2015 and 2016 in East Region by State')
plt.show()
```

# In East Region, NY registered a decline Sales in 2016

```python
#5
hightier=sales_data.loc[sales_data.Tier=='High']
hightier
```

```python
hightier=hightier.groupby(by='Division')[['Units2015','Units2016']].sum()
hightier.plot(kind='bar',figsize=[20,10],title='Comparison of Units sold in 2015 and 2016 in High Tier by Division')
plt.show()
```

# We observe that in High Tier, none of Division is declining in number of units sold in 2016 compared to 2015

```python
#6
Qtr=[ ]
for x in sales_data.Month:
    if x in ['Jan','Feb','Mar']:
        Qtr.append('Q1')
    elif x in ['Apr','May','Jun']:
        Qtr.append('Q2')
    elif x in ['Jul','Aug','Sep']:
        Qtr.append('Q3')
    else:
        Qtr.append('Q4')
        
```

```python
Qtr
```

```python
sales_data['Qtr']=pd.Series(Qtr)
```

```python
sales_data
```

```python
#7
qtr_sales=sales_data.groupby(by='Qtr')[['Sales2015','Sales2016']].sum()
qtr_sales.plot(kind='bar',title='Qtr wise Sales in 2015 and 2016')
plt.show()
```

```python
#8
qtr_tier_sales=sales_data.groupby(by=['Qtr','Tier'])['Sales2016'].sum()
qtr_tier_sales
```

```python
#Q1
Q1=qtr_tier_sales.loc['Q1',:]
Q2=qtr_tier_sales.loc['Q2',:]
Q3=qtr_tier_sales.loc['Q3',:]
Q4=qtr_tier_sales.loc['Q4',:]




```

```python
#dropping the index 'OUT' because it contains negative value and not allowing to plot the pie chart
Q3=qtr_tier_sales.loc['Q3',:] 
Q3=Q3.drop(index='Out')
Q3
```

```python
Q1.plot(kind='pie',autopct="%1.0f%%",title='Q1')
plt.show()
Q2.plot(kind='pie',autopct="%1.0f%%",title='Q2')
plt.show()
Q3.plot(kind='pie',autopct="%1.0f%%",title='Q3')
plt.show()
```
